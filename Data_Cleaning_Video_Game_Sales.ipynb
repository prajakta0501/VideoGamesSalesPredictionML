```python
# Replaced tbd values with None
#shows count of user score
df_games.User_Score.replace('tbd', None, inplace = True)
df_games.User_Score.value_counts()
```

```python
# change the type
df_games['User_Score'] = df_games['User_Score'].astype(float)
df_games.info()
```

```python
# missing values
df_games.isna().sum()
```

```python
df_games.columns = map(str.lower,  df_games.columns)
df_games.info()
```

```python
df_games.groupby(df_games['rating']).count()
```

```python
# few of this ratings have few games. So need to drop them
df_games[df_games['rating'].isin(['AO','EC','K-A','RP'])]
```

```python
# using logical_not to invert it
mask = np.logical_not(df_games['rating'].isin(['AO','EC','K-A', 'RP']))
# filtering out those unwanted games
df_games= df_games[mask]
# sanity check
df_games['rating'].unique()
```

```python
arr_mode=['name','genre','publisher','developer','rating']
for col in arr_mode:
    df_games[col] = df_games[col].fillna(df_games[col].mode()[0])   
```

```python
df_games['year_of_release'] = df_games['year_of_release'].fillna(df_games['year_of_release'].median())
```

```python
arr_mean=['na_sales','eu_sales','jp_sales','other_sales','global_sales','user_score','user_count','critic_score','critic_count']
for col in arr_mean:
    df_games[col] = df_games[col].fillna(df_games[col].mean())  
```

```python
df_games.isna().sum()
```

```python
arr_salesinmillions=['na_sales','eu_sales','jp_sales','other_sales','global_sales'] 

for sal in arr_salesinmillions:
    df_games[sal]=df_games[sal]
```
